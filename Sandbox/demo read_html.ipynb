#### Get the first dataframe

```python
df = pd.read_html(url, thousands = '.', decimal = ',')[0]
df
```

```python
df.rename(columns = {'Bevolking[1]': 'Bevolking', 
                     'Oppervlakte(land+water) in km²':'Oppervlakte',
                     'Oppervlakte(land) in km²':'Oppervlakte Land',
                     'inw./km²[2]':'inwonersdichtheid'}, 
          inplace = True)
```

```python
df['Provincie'] = df['Provincie'].astype('string')
df['Hoofdstad'] = df['Hoofdstad'].astype('string')
df['Gemeente met demeeste inwoners'] = df['Gemeente met demeeste inwoners'].astype('string')
```
